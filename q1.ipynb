```python
import pandas as pd
import numpy as np

# Creating dictionary with null values
data = {
    "Student_Name": ["Sanskar", "Riya", "Kunal", "Sneha", "Rohit", "Neha"],
    "Semester": ["Sem 1", "Sem 2", "Sem 1", "Sem 3", "Sem 2", "Sem 1"],
    "Marks": [78, 85, 92, 30, 88, np.nan],   # null value in marks
    "Height": [165, 158, 172, 160, np.nan, 155],  # null value in height
    "Age": [19, 20, 18, 21, 20, np.nan],     # null value in age
    "Study_Hours": [5, 6, 8, 2, 7, 4]
}

# Creating DataFrame
df = pd.DataFrame(data)

print(df)
```

```python
semester_count = df["Semester"].value_counts()
print(semester_count)
```

```python
# Remove null values for outlier detection
marks = df["Marks"].dropna()

Q1 = marks.quantile(0.25)
Q3 = marks.quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df["Marks"] < lower_bound) | (df["Marks"] > upper_bound)]

print("Outliers in Marks column:")
print(outliers)
```

```python
correlation = df["Marks"].corr(df["Study_Hours"])
print("Correlation between Marks and Study Hours:", correlation)
```
